### Importing data into SQLite

```python
import sqlite3

def run_query(q):
    with sqlite3.connect('mlb.db') as conn:
        return pd.read_sql(q, conn)

def run_command(c):
    with sqlite3.connect('mlb.db') as conn:
        conn.execute('PRAGMA foreign_keys = ON;')
        conn.isolation_level = None
        conn.execute(c) 
def show_tables():
    
    q = "SELECT name, type FROM sqlite_master WHERE type IN ('table','view');"
    return run_query(q)


# Open connection to database
conn = sqlite3.connect('mlb.db')

# Create database tables
game_log.to_sql('game_log', conn, index=False, if_exists='replace')

park_codes.to_sql('park_codes', conn, index=False, if_exists='replace')

person_codes.to_sql('person_codes', conn, index=False, if_exists='replace')

team_codes.to_sql('team_codes', conn, index=False, if_exists='replace')

# Show tables in database
show_tables()

# Add column to game_log table

c1 = """
        ALTER TABLE game_log
        ADD COLUMN game_id TEXT;
    """
try:
    run_command(c1)
except:
    pass
c2 = """
        UPDATE game_log
        SET game_id = date || h_name || number_of_game
        WHERE game_id IS NULL;
     """
try:
    run_command(c2)
except:
    pass

#check that values for game_id are being created correctly
q = """
        SELECT 
            game_id,
            date,
            h_name,
            number_of_game
        FROM game_log
        LIMIT 10;
    """

run_query(q)
```

### Normalization Opportunities:

```python
# Create person table
c3="""
    CREATE TABLE IF NOT EXISTS person (
        person_id TEXT PRIMARY KEY,
        first_name TEXT,
        last_name TEXT
    );
    """
try:
    run_command(c3)
except:
    pass

#Insert values into person table
c="""
    INSERT OR IGNORE INTO person 
    SELECT
        id, 
        first, 
        last
    FROM person_codes;
    """
try:
    run_command(c)
except:
    pass


#output first 5 rows of table
q="""
    SELECT * FROM person LIMIT 5;
"""

run_query(q)
#show_tables()
```
